```python
import numpy as np
import matplotlib.pyplot as plt
x_decode = np.load("decode_x.npy")
dx_decode = np.load("dx_decode.npy")
ddx_decode = np.load("ddx_decode.npy")
```

```python
x_decode = x_decode.reshape(20,27,24)
dx_decode = dx_decode.reshape(20,27,24)
ddx_decode = ddx_decode.reshape(20,27,24)
```

```python
for i in range(20):
    img = x_decode[i*4]
    plt.imshow(img, interpolation='nearest')
    plt.colorbar()
    plt.show()
```

```python
for i in range(20):
    img = dx_decode[i]
    plt.imshow(img, interpolation='nearest')
    plt.colorbar()
    plt.show()
```

```python
for i in range(20):
    img = ddx_decode[i]
    plt.imshow(img, interpolation='nearest')
    plt.colorbar()
    plt.show()
```
